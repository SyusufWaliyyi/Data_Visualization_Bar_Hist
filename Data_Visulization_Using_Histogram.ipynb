```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
height_df = pd.read_csv('height.csv')
# Display the first few rows of the DataFrame
height_df.head()
```

Bar plot base on gender

```python
fig, ax = plt.subplots()
sns.barplot(x='Gender', y='Hight', data=height_df, ax=ax, hue='Gender', palette='Set2')
ax.set_title('Height Distribution by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Height (cm)')
ax.grid(axis='y')
plt.show()
```

```python
gender_count = height_df.groupby('Gender')['Gender'].count()
gender_count
```

```python
# Male data frame
male_df = height_df[height_df['Gender'] == 'M']
male_df.head()
```

```python
# Women data frame
female_df = height_df[height_df['Gender'] == 'F']
female_df.head()
```

```python
fig, ax = plt.subplots()
ax.hist(male_df['Hight'], bins=20, color='blue', alpha=0.7)
ax.set_title('Male Height Distribution')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Frequency')
ax.grid(axis='y', alpha=0.75)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(male_df['Weight'], bins=20, color='blue', alpha=0.7)
ax.set_title('Male Weight Distribution')
ax.set_xlabel('Weight (Kg)')
ax.set_ylabel('Frequency')
ax.grid(axis='y', alpha=0.75)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(male_df['Shoe size'], bins=20, color='blue', alpha=0.7)
ax.set_title('Male Shoe size Distribution')
ax.set_xlabel('Shoe size')
ax.set_ylabel('Frequency')
ax.grid(axis='y', alpha=0.75)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(female_df['Hight'], bins=20, color='blue', alpha=0.7)
ax.set_title('Female Height Distribution')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Frequency')
ax.grid(axis='y', alpha=0.75)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(female_df['Weight'], bins=20, color='blue', alpha=0.7)
ax.set_title('Female Weight Distribution')
ax.set_xlabel('Weight (Kg)')
ax.set_ylabel('Frequency')
ax.grid(axis='y', alpha=0.75)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(female_df['Shoe size'], bins=20, color='blue', alpha=0.7)
ax.set_title('Female Shoe size Distribution')
ax.set_xlabel('Shoe size')
ax.set_ylabel('Frequency')
ax.grid(axis='y', alpha=0.75)
plt.show()
```

Combine plots

```python
fig, ax = plt.subplots()
ax.hist(male_df['Hight'], bins=20, color='green', alpha=0.5)
ax.hist(female_df['Hight'], bins=20, color='red', alpha=0.5)
ax.set_title('Height Distribution')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Frequency')
ax.legend(['Male', 'Female'])
ax.grid(axis='y', alpha=0.75)
```

```python
fig, ax = plt.subplots()
ax.hist(male_df['Weight'], bins=20, color='green', alpha=0.5, histtype='step')
ax.hist(female_df['Weight'], bins=20, color='red', alpha=0.5, histtype='step')
ax.set_title('Weight Distribution')
ax.set_xlabel('Weight (Kg)')
ax.set_ylabel('Frequency')
ax.legend(['Male', 'Female'])
ax.grid(axis='y', alpha=0.75)
```

```python
fig, ax = plt.subplots()
ax.hist(male_df['Shoe size'], bins=20, color='green', alpha=0.7)
ax.hist(female_df['Shoe size'], bins=20, color='red', alpha=0.7)
ax.set_title('Shoe size Distribution')
ax.set_xlabel('Shoe size')
ax.set_ylabel('Frequency')
ax.legend(['Male', 'Female'])
ax.grid(axis='y', alpha=0.75)
```
