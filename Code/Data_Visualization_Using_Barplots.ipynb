```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
olym = {
    'Country': ['USA', 'China', 'Russia', 'Great Britain', 'Germany', 'France', 'Italy', 'Japan', 'South Korea', 'Australia'],
    'Gold': [2654, 2242, 1950, 850, 1700, 1500, 800, 600, 500, 400],
    'Silver': [1064, 1000, 800, 600, 500, 400, 300, 300, 200, 100],
    'Bronze': [1000, 800, 600, 400, 300, 200, 100, 100, 100, 100],
    'Total': [4212, 3224, 1950, 1700, 1700, 1500, 800, 600, 500, 400]
}

# Create the DataFrame
olympics_df = pd.DataFrame(olym)

# Set the index to 'Country'
olympics_df = olympics_df.set_index('Country')
olympics_df
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(olympics_df.index, olympics_df['Gold'], color='blue')
ax.set_title('Olympic Gold Medals by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Number of Gold Medals')
ax.set_xticklabels(olympics_df.index, rotation=30)
# Add grid lines for better readability
ax.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(olympics_df.index, olympics_df['Silver'], color='green')
ax.set_title('Olympic Silver Medals by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Number of Silver Medals')
ax.set_xticklabels(olympics_df.index, rotation=30)
# Add grid lines for better readability
ax.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(olympics_df.index, olympics_df['Bronze'], color='red')
ax.set_title('Olympic Bronze Medals by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Number of Bronze Medals')
ax.set_xticklabels(olympics_df.index, rotation=30)
# Add grid lines for better readability
ax.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(olympics_df.index, olympics_df['Gold'], color='blue')
ax.bar(olympics_df.index, olympics_df['Silver'], bottom=olympics_df['Gold'], color='green')    
ax.bar(olympics_df.index, olympics_df['Bronze'], bottom=olympics_df['Gold'] + olympics_df['Silver'], color='red')
ax.set_title('Olympic Medals by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Number of Medals')
ax.set_xticklabels(olympics_df.index, rotation=30)
# Add grid lines for better readability
ax.grid(True)
# add legend
ax.legend(['Gold', 'Silver', 'Bronze'], loc='upper right')
plt.show()
```
