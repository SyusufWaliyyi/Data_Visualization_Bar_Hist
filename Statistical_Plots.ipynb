```python
# import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('default')
```

```python
height_df = pd.read_csv('height.csv')
# Display the first few rows of the DataFrame
height_df.head()
```

```python
male_df = height_df[height_df['Gender'] == 'M']
female_df = height_df[height_df['Gender'] == 'F']
```

```python
fig, ax = plt.subplots()
# Create a error bar plot
ax.bar('Male', male_df['Hight'].mean(), 
       yerr=male_df['Hight'].std(), color='green')
ax.bar('Female', female_df['Hight'].mean(), 
       yerr=female_df['Hight'].std(), color='red')
ax.set_ylabel('Height (cm)')
ax.set_title('Average Height with Standard Deviation')
ax.grid()
# Show the plot
plt.show()

fig.savefig('height_bar_plot.png', dpi=300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
# Create an error bar plot
ax.bar('Male', male_df['Hight'].mean(), 
       yerr=male_df['Hight'].sem())
ax.bar('Female', female_df['Hight'].mean(), 
       yerr=female_df['Hight'].sem())
ax.set_ylabel('Height (cm)')
ax.set_title('Average Height with Standard Error')
ax.grid()
# Show the plot
plt.show()

fig.savefig('height_Error_bar_plot.png', dpi=300, bbox_inches='tight')
```

Adding boxplot

```python
fig, ax = plt.subplots()
# Create a boxplot for height by gender
ax.boxplot([male_df['Hight'], female_df['Hight']])
ax.set_xticklabels(['Male', 'Female'])
ax.set_ylabel('Height (cm)')
ax.set_title('Height by Gender')
ax.grid()
# Show the plot
plt.show()

fig.savefig('height_box_plot.png', dpi=300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots()
# Create a boxplot for height by gender
ax.boxplot([male_df['Weight'], female_df['Weight']])
ax.set_xticklabels(['Male', 'Female'])
ax.set_ylabel('Weight (Kg)')
ax.set_title('Weight by Gender')
ax.grid()
# Show the plot
plt.show()

fig.savefig('weight_box_plot.png', dpi=300, bbox_inches='tight')
```

Scatter plot

```python
fig, ax = plt.subplots()
# Create a scatter plot for height vs weight
ax.scatter(male_df['Hight'], male_df['Weight'])
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Weight (Kg)')
ax.set_title('Height vs Weight')
ax.grid()
# Show the plot
plt.show()

fig.savefig('height_Weight_scatter_plot.png', dpi=300, bbox_inches='tight')
```

```python
# scatter base on gender
fig, ax = plt.subplots()
# Create a scatter plot for height vs weight
ax.scatter(male_df['Hight'], male_df['Weight'], 
           label='Male', color='green')
ax.scatter(female_df['Hight'], female_df['Weight'], 
           label='Female', color='red')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Weight (Kg)')
ax.set_title('Height vs Weight')
ax.grid()
ax.legend()
# Show the plot
plt.show()

fig.savefig('height_weight_scatter_plot.png', dpi=300, bbox_inches='tight')
```
